```python
import collections
Card = collections.namedtuple("Card", ['rank','suit'])
```

```python
Card.__doc__
```

- 可以认为类的命名和变量的命名与java规范一致
- __len__ 和 __getitem__ 即可像操作普通list一样操作自定义类，collection中有详细实现
- 明确一点，**特殊方法**是给python解释器调用的，而不是自己
- Python内置类型，例如list、str、bytearray、Numpy数组等扩展，Python解释器会抄个近路，直接使用底层c结构体中的长度属性
- 特殊方法的调用很多时候是隐式调用的，比如for i in x: 其实式在背后调用iter(x)，然后调用x.__iter__()或者x.__getitem__()



```python
class FrenchDeck:
    ranks = [str(n) for n in range(2,11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self) :
        self._cards = [Card(rank, suit) for suit in self.suits
                      for rank in self.ranks]
        
    def __len__(self):
        return len(self._cards)
    
    def __getitem__(self, position):
        return self._cards[position]
        
```

```python
suit_values=dict(spades=3, diamonds=2, clubs=1, hearts=0)

def spades_high(card):
    rank_value = FrenchDeck.ranks.index(card.rank)
    return rank_value*len(suit_values) + suit_values[card.suit]

for card in sorted(deck, key=spades_high):
    print(card)
```
